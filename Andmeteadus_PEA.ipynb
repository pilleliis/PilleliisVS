```python
andmed_kl=andmed[["2021LifeExpectancyAtBirth","2021GDPpercapita" ]]   # teeme uue tabeli kahe veeru andmete põhjal, et saaks andmeid väärtuste põhjal klasterdada

andmed_kl.columns=["Oodatav_eluiga", "SKP"]  #muudame veeru nime, mis hiljem hakkasid numbri tõttu streikima


andmed_kl = andmed_kl.dropna(subset=['Oodatav_eluiga'])   #eemaldame read, kus ei ole 2021LifeExpectancyAtBirth märgitud - KMeans ei söö NaN-e
andmed_kl.info()
andmed_kl.head()
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
andmed_stand=StandardScaler().fit_transform(andmed_kl)   #Kuna andmestikus on väärtused väga erinevatel skaaladel, siis tuleb need standardiseerida
```

```python
#visualiseerime standardiseeritud andmed
plt.subplots(figsize=(6, 6))
plt.scatter(andmed_stand.T[0], andmed_stand.T[1])

plt.xlabel("Oodatav_eluiga")
plt.ylabel("SKP")
plt.title("Oodatava eluea ja SKP seos");
```

```python

from sklearn.cluster import KMeans   #K-Means tegelikult eeldab, et andmete vahel ei ole korrelatsiooni ja siis on klasterdamisel mõtet, aga mängime siiski läbi.
km=KMeans(n_clusters=3, max_iter=50, n_init=10)   #Proovisime ka 6 klastriga, aga seda oli liiga palju, 3 on selgemad klastrid
fitted_km=km.fit(andmed_stand)   #rakendame algoritmi meie andmetele. Tulemuseks KMeans objekt, mis muuhulgas sisaldab klastrite tsentrite koordinaate
```

```python
#tsentrite koordinaadid
fitted_km.cluster_centers_
```

```python
#visualiseerime graafikul ka tsentrid
plt.subplots(figsize=(6, 6))

#joonistame kõik andmepunktid
plt.scatter(x=andmed_stand[:, 0], y=andmed_stand[:, 1]);


#sns.scatterplot(x = andmed_stand.Oodatav_eluiga,
               #y= andmed_stand.SKP,
              # hue=andmed_stand.cluster,
               #palette={0:'yellow', 1:'green', 2:'black'})

#märgime tsentrid
plt.scatter(fitted_km.cluster_centers_[:, 0], fitted_km.cluster_centers_[:, 1], marker='*', s=300,
            color='black', label='centroid');

plt.xlabel("Oodatav_eluiga")
plt.ylabel("SKP")
plt.title("Oodatava eluea ja SKP seos");
```

```python
predictions=km.predict(andmed_stand)  #kasutame mudelit, et ennustada klastreid. Iga riik saab klastri tunnuse
predictions
```

```python
andmed_kl["cluster"] = predictions  #teeme tabelisse uue veeru, kus on klastri väärtus. Seega on meil riigid jagatud 3 klastrisse oodatava eluea ja SKP järgi.
andmed_kl["Country Name"] = andmed["Country Name"]  #kuna tegutsesime väiksema tabeliga, kus oli ainutl 2 veergu, siis lisame ka riikide nimede veeru.
andmed_kl["Region"] = andmed["Region"]   #...ja regioni veeru
# Filtrime andmed, et kuvada ainult klastri väärtusega 1
andmed_kl_cluster_1 = andmed_kl[andmed_kl["cluster"] == 1]
#print(andmed_kl_cluster_1)
andmed_kl_cluster_1.Region.value_counts()   #Klastrisse 1 kuuluvad põhiliselt Aafrika riigid ning ka Ida-Aasiat. KÕige madalam SKP ning kõige väiksem oodatav eluiga
```

```python
andmed_kl_cluster_1.Oodatav_eluiga.describe()  #Kõige nõrgema klastri oodatav eluiga max väärus on 68aastat ning SKP elaniku kohta 11 488, mis on üsna väike.
#EHk kui sünnid keskmise SKP väärtusega 2300$ elaniku kohta riiki, siis keskmine eluiga on u 62 aastat.
```

```python
andmed_kl_cluster_1.SKP.describe()
```

```python
# Filtrime andmed, et kuvada ainult klastri väärtusega 0
andmed_kl_cluster_2 = andmed_kl[andmed_kl["cluster"] == 0]
#print(andmed_kl_cluster_2)
andmed_kl_cluster_2.Region.value_counts()   #Klastrisse 0 kuuluvad põhiliselt Ladina-Ameerikat, Euroopat. Ida-Aasiat ning ka Põhja-Aafrikat
```

```python
# Filtrime andmed, et kuvada ainult klastri väärtusega 2
andmed_kl_cluster_3 = andmed_kl[andmed_kl["cluster"] == 2]
#print(andmed_kl_cluster_3)
andmed_kl_cluster_3.Region.value_counts()   #Klastrisse 2 kuuluvad põhiliselt jõukam Eurpoopa, Aasia jõukam osa nagu nt Singapur, Korea, Jaapan.  Ilmselgelt on jõukad riigi (kõrge SKPga) ka kõrgem aoodatava elueaga.
```

```python
#proovime kõiki olulisi tunnuseid koos klasterdada - selleks loome uue tabeli, millega klasterdada
#valime veerud, mis on kohased/vajalikud analüüsiks, mida saab numbrile taandada
klasterandmed=andmed[["Region", "2021LifeExpectancyAtBirth","2021GDPpercapita", "CO2 Emissions in tons per capita", "Population density per square km", "Fertility rate", "2021InfantDeatsPer1000", "Access%toElectricity",  ]]
#meil on andmestikus veerus 2021InfantDeatsPer1000 21 null väärtust. Seda on liiga palju, et ridu kustutada. Seega täidame tühjad kohad vastava regiooni keskmisega. Sama teeme oodatava elueaga.
klasterandmed.loc[:, '2021InfantDeatsPer1000'] = klasterandmed.groupby('Region')['2021InfantDeatsPer1000'].transform(lambda x: x.fillna(x.mean()))
klasterandmed.loc[:,'2021LifeExpectancyAtBirth'] = klasterandmed.groupby('Region')['2021LifeExpectancyAtBirth'].transform(lambda x: x.fillna(x.mean()))
klasterandmed.isnull().sum()
```

```python
#Regiooni veergu oli vaja uute tunnuste loomiseks, kustutame, sest tekstiveerge ei saa klasterdada.
klasterandmed=klasterandmed.drop(["Region"], axis=1)
klasterandmed.info()
```

```python
km = KMeans(n_clusters=4, max_iter=50, n_init=10)
fitted_km = km.fit(klasterandmed)
predictions = km.predict(klasterandmed)
klasterandmed["cluster"] = predictions
klasterandmed.head()
```

```python
centroids_2d = pca.transform(km.cluster_centers_)

fig, ax = plt.subplots(figsize=(6, 6))

plt.scatter(centroids_2d[:, 0], centroids_2d[:, 1], marker='*', s=300,
            color='b');

colors = ["red", "white", "yellow", "green", "orange", "black"]
for i in range(5):
    plt.scatter(pca_2d[klasterandmed["cluster"] == i, 0], pca_2d[klasterandmed["cluster"] == i, 1], color=colors[i]);

plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Principal components");
```

```python
#Evelini katsetused
#Oodatav eluiga vs imikute suremus
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('Andmestik.csv')
df_filtered = df[df['Status_Country1_Region0'] == 1]
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='2021LifeExpectancyAtBirth', y='2021InfantDeatsPer1000', size='Population', hue='Region', palette='coolwarm', sizes=(20, 200))
sns.regplot(data=df, x='2021LifeExpectancyAtBirth', y='2021InfantDeatsPer1000', scatter=False, color='black', line_kws={"linewidth": 1})
plt.title('Life Expectancy vs. Infant Mortality')
plt.xlabel('Life Expectancy at Birth (years)')
plt.ylabel('Infant Deaths per 1000 live births')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#Evelini katsetused
#oodatav eluiga vs CO2 per capita


bins = [0, 1, 5, 10, 20, 50, 100, 200]
labels = ['0-1', '1-5', '5-10', '10-20', '20-50', '50-100', '100-200']
df_filtered['CO2 Emissions per capita bins'] = pd.cut(df_filtered['CO2 Emissions in tons per capita'], bins=bins, labels=labels)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_filtered, x='2021LifeExpectancyAtBirth', y='CO2 Emissions in tons per capita', hue='CO2 Emissions per capita bins', palette='viridis', size='Population', sizes=(20, 200))


sns.regplot(data=df_filtered, x='2021LifeExpectancyAtBirth', y='CO2 Emissions in tons per capita', scatter=False, color='black', line_kws={"linewidth": 1})

plt.title('Life Expectancy at Birth vs. CO2 Emissions per Capita')
plt.xlabel('Life Expectancy at Birth (years)')
plt.ylabel('CO2 Emissions per Capita (tons)')
plt.legend(title='CO2 Emissions per Capita Bins', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
